# Heatmap

A heatmap is a graphical representation of data where the individual values contained in a matrix are represented as colors. It is a bit like looking a data table from above.

A heatmap is also useful to display the result of hierarchical clustering. Basically, clustering checks which countries tend to have the same features on their numeric variables, and therefore which countries are similar. The usual way to represent the result is to use dendrograms.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### Heatmap Seaborn

- **data:** Matrix of data to visualize.
- **annot:** If True, displays the values within each cell.
- **fmt:** Format of the values (.2f for two decimals, d for integers, etc.).
- **cmap:** Color palette (coolwarm, viridis, magma, etc.).
- **linewidths:** Thickness of the lines between cells.
- **linecolor:** Color of the lines between cells.
- **cbar:** If True, displays the color bar.
- **xticklabels, yticklabels:** Axis labels (True, False, or custom list).

```python
np.random.seed(10)
data = np.random.rand(5, 5)

# Heatmap con Seaborn
sns.heatmap(data, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
```
