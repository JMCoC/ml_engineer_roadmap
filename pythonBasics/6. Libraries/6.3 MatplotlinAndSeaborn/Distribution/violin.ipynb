# Violin

Violinplots allow to visualize the distribution of a numeric variable for one or several groups. It is really close from a boxplot, but allows a deeper understanding of the distribution.

Violins are particularly adapted when the amount of data is huge and showing individual observations gets impossible.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../../CSV/customer_shopping_data.csv")
```

```python
df.head()
```

### Violin Seaborn
- **data:** Dataframe, Series, dict, array or list of arrays
- **x, y, hue:** names of variables in data
- **split:** if **true**, show an un-mirrored distribution
- **inner_kws:** Keyword arguments for the “inner” plot, passed to one of:
    - stick, box, quartile, point
- **cut:** Distance, in units of bandwidth, to extend the density past extreme datapoints. Set to 0 to limit the violin within the data range.

```python
sns.violinplot(data=df, x='gender', y='price')
plt.show()
```

### Violin Matplotlib
- **dataset:** Array or a sequence of vector
- **showmeans, showmedians, showextrema:** if **True** shows the mean, median, min/max
- **bw_method:** {'scott', 'silverman'} or float or callable, default: 'scott',    The method used to calculate the estimator bandwidth

```python
fig, ax = plt.subplots()
ax.violinplot([df[df['gender'] == 'Male']['price'], df[df['gender'] == 'Female']['price']], showmeans=False, showmedians=True)
ax.set_xticks([1, 2])
ax.set_xticklabels(['Male', 'Female'])
ax.set_xlabel('Gender')
ax.set_ylabel('Price')
plt.show()
```
