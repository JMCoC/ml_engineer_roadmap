# Histogram

A histogram is an accurate graphical representation of the distribution of a numeric variable. It takes as input numeric variables only. The variable is cut into several bins, and the number of observation per bin is represented by the height of the bar.

Several distribution shapes exist, here is an illustration of the most common ones:
1. Uniform Distribution: All outcomes are equally likely; the probability of each outcome is the same.
2. Normal Distribution: A bell-shaped curve, where the mean, median, and mode are all equal.
3. Bimoal Distribution: Two peaks, indicating two different modes.
4. Comb Distribution: The bars are alternating between high and low values.
5. Edge peak: The peak is at the edge of the distribution.
6. Skewed Distribution: The data distribution is not symmetrical, it falls more to one side than the other.
7. Uniform Distribution: All outcomes are equally likely; the probability of each outcome is the same.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../../CSV/customer_shopping_data.csv")
```

```python
df.head()
```

### Hist Seaborn
- **data:** Dataframe, numpy array, mapping, sequence
- **x, y, hue:** names of variables in data
- **bins:** str, number, vector, or a pair of such values. Generic bin parameter that can be the name of a reference rule, the number of bins, or the breaks of the bins
- **binwidth:** Width of each bin
- **multiple:** Approach to resolving multiple elements when semantic mapping creates subsets. Only relevant with univariate data. (Layer, stack, fill)
- **element:** {“bars”, “step”, “poly”} Visual representation of the histogram statistic. Only relevant with univariate data.
- **kde:** If **True**, compute a kernel density estimate to smooth the distribution and show on the plot as (one or more) line(s).

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True, color='red')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

### Matplotlib Hist

- **x:** data
- **bins:** str, number, sequence. Default 10
- **density:** If **True**, each bin will display the bin's raw count divided by the total number of counts and the bin width
- **cumulative:** If **True**, then a histogram is computed where each bin gives the counts in that bin plus all bins for smaller values.
- **histtype:** {'bar', 'barstacked', 'step', 'stepfilled'}, default: 'bar'
The type of histogram to draw.
- **color:** color or list of color or None, default: None

```python
plt.hist(df['price'], bins=30, color='red')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```
