# Density

A density plot shows the distribution of a numeric variable. It takes only numeric variables as input and is very close from an histogram. It can be use in the same exact condition.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../../CSV/customer_shopping_data.csv")
```

```python
df.head()
```

### Kernel density estimate (KDE) Seaborn
- **data:** Dataframe, numpy array, mapping, sequence
- **x, y, hue:** names of variables in data
- **bw_adjust:** Factor that multiplicatively scales the value chosen using    bw_method. Increasing will make the curve smoother.
- **cumulative:** if **True**, estimate a cumulative distribution function. Requires scipy. 
- **multiple:** Method for drawing multiple elements when semantic mapping creates subsets. Only relevant with univariate data. (Layer, stack, fill)
- **common_norm:** If **True**, scale each conditional density by the number of observations such that the total area under all densities sums to 1

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='age', bw_adjust=0.5, hue='category', fill=True)
plt.title('Density Plot of Price')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

- Although Matplotlib does not have a specific function for density plots, the `hist()` function with the parameter `density=True` can be used to estimate the probability density function.
